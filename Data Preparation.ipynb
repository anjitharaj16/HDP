# Early Heart Disease Prediction Detection Using Machine Learning Algorithms

## Data Preparation

```python
# importing all the necessary libraries and classes
import pandas as edhd_pd
import matplotlib.pyplot as edhd_plot
import missingno as edhd_msno
import seaborn as edhd_sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# loading ckd dataset

edhd_DataFrame= edhd_pd.read_csv('EDHD_initial_dataset.csv')
edhd_DataFrame
```

```python
# Converting the categorical column values
edhd_DataFrame['sex'] = edhd_DataFrame['sex'].replace({0: 'female', 1: 'male'})
edhd_DataFrame['cp'] = edhd_DataFrame['cp'].replace({0: 'typical angina', 1: 'atypical angina', 2: 'non-anginal pain', 3: 'asymptomatic'})
edhd_DataFrame['restecg'] = edhd_DataFrame['restecg'].replace({0: 'normal', 1: 'abnormality',3:'positive'})
edhd_DataFrame['fbs'] = edhd_DataFrame['fbs'].replace({0: 'Yes', 1: 'No'})
edhd_DataFrame
```

```python
edhd_DataFrame.to_csv('EDHD_initial_dataset1.csv', index=False)
```
