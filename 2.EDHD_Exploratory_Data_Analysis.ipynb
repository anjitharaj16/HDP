```python
#numerical feature distribution
numerical_columns = [featuredColumn for featuredColumn in edhd_DataFrame.columns if edhd_DataFrame[featuredColumn].dtype != 'object']
edhd_pd.set_option('display.max_columns', 26)
edhd_plot.style.use('fivethirtyeight')

edhd_plot.figure(figsize = (20, 20))
plotCount = 1
```

```python
#numerical feature distribution
numerical_columns = [featuredColumn for featuredColumn in edhd_DataFrame.columns if edhd_DataFrame[featuredColumn].dtype != 'object']
edhd_pd.set_option('display.max_columns', 14)
edhd_plot.style.use('fivethirtyeight')

edhd_plot.figure(figsize = (20, 20))
plotCount = 1

for featuredColumn in numerical_columns:
    if plotCount <= len(numerical_columns):
        plot_axis = edhd_plot.subplot(5, 4,plotCount)
        edhd_sns.distplot(edhd_DataFrame[featuredColumn], kde=True, color='green')  # Adjust color here)
        edhd_plot.xlabel(featuredColumn)
    plotCount+=1


edhd_plot.show()
```

```python
 #categorical columns label distribution
categorical_columns = [featuredColumn for featuredColumn in edhd_DataFrame.columns if edhd_DataFrame[featuredColumn].dtype == 'object']

edhd_plot.figure(figsize = (20, 10))
plotCount = 1


for featuredColumn in categorical_columns:
    if plotCount <= len(categorical_columns):
        plot_axis = edhd_plot.subplot(2, 2, plotCount)
        edhd_sns.countplot(data=edhd_DataFrame, x=featuredColumn)
        edhd_plot.xlabel(featuredColumn)

    plotCount += 1

edhd_plot.show()
```

```python
#Draw correlation matrix to find any corraltion 
# Compute the correlation matrix
corrmat = edhd_DataFrame.corr(numeric_only=True)

# Set the features to be displayed on the heatmap
top_corr_features = corrmat.index

# Set the size of the figure
edhd_plot.figure(figsize=(20,10))

# Plot heatmap
edhd_sns.heatmap(corrmat, annot=True, cmap="YlOrRd")

edhd_plot.savefig('heatmap.png', dpi=300, bbox_inches='tight')


# Display the heatmap
edhd_plot.show()
```

```python


# Create a new figure with a larger size for better readability
edhd_plot.figure(figsize=(12, 8))

# Plot the boxplot
edhd_DataFrame[numerical_columns].boxplot(grid=True)

# Add titles and labels for clarity
edhd_plot.title('Box Plot of Numerical Features')
edhd_plot.ylabel('Value')
edhd_plot.xlabel('Features')

# Rotate x-axis labels for better readability
edhd_plot.xticks(rotation=90)

# Save the figure to a file
edhd_plot.savefig('box plots.png', dpi=300, bbox_inches='tight')

# Display the plot
edhd_plot.show()
```
